**Subscriber-count Comparision Between Two Youtube Gaming Channels** 


This notebook proposes a python based solution to find subscriber count of two Youtube channels. The goal is to find subscriber difference between two channels that covers a very similar topic. Specifically, two gaming related channels named, "Super Bunnyhop" and "Jim Sterling" have been selected.

Both of the mentioned channels cover major gaming related news,  reviews of games and opinions on current gaming trends in terms of game designs and also about business practices of the publishers. Their major focus on the channels is providing well-researched journalism on gaming media that is severely lacking on other websites and youtube channels. 

There are three major differences between the two channels. Firstly, "Jim Sterling" produces new videos every week whereas "Super Bunnyhop" usually produces new view bi-weekly. "Jim Sterling" also previews newly released along with new related videos, on the other hand, "Super Bunnyhop" never previews any games, this channel mostly focuses on news related videos and occasionally produces lengthy critique about various games. The last major difference between them is their presentation. "Jim Sterling" fills his videos with jokes, uses a stage persona for presentation where "Super Bunnyhop" lean towards a generic way of presenting news without any filler segments with jokes or skits.  

Therefore, these two channels have been chosen for there somewhat similar approach towards gaming journalism. The subscriber difference will shed light to the fact which approach of providing gaming news, critique is working the best and reaching more audience.

Now, we start writing the code for scarping the YouTube channels to find out the subscriber difference. 

To scrape the Youtube channels to find the subscriber count channels, Requests and BeautifulSoup libraries have been used. 
First, let's start with importing these libraries.

```
import requests
from bs4 import BeautifulSoup
import re

#following two libraries are needed for importing image files into the notebook
from google.colab import files 
from IPython.display import Image 
```

Now, we have to save the website links of the YouTube channels we are interested in for scraping subscriber data. Specifically, we are scraping the "About" page of the channels as the general information can be found on that page.

```
jim_sterling = 'https://www.youtube.com/user/JimSterling/about'
super_bunny_hop = 'https://www.youtube.com/user/bunnyhopshow/about'
```

```
req_jim = requests.get(jim_sterling)
req_sbh = requests.get(super_bunny_hop)
```

```
jimSoup = BeautifulSoup(req_jim.text, "html.parser")
sbhSoup = BeautifulSoup(req_sbh.text, "html.parser")
```

```
print(jimSoup)
```

```
print(sbhSoup)
```

```
upload = files.upload() # uploading image
```

To find the subscriber information, the "About" pages in the Youtube channels, 'https://www.youtube.com/user/JimSterling/about' and 'https://www.youtube.com/user/bunnyhopshow/about' respectively, have been parsed. From there the tag and class elements can be found which are "span" and "about-stat" respectively. In the class "about-stat", there is multiple information about the channels but as our concern is with the subscription numbers, only the subscription data will be printed. Screenshots of the elements in the two channels are shown below. 

```
Image('cbh.PNG') #Tag elemets of "Super Bunnyhop"
```

```
Image('jim.PNG') # Tag elements of "Jim Sterling"
```

```
jimTag = jimSoup.findAll("span", {"class":"about-stat"})
sbhTag = sbhSoup.findAll("span", {"class":"about-stat"})
```

After the parsing of the two pages, we can find the subscriber count of each Youtube Channels. 

```
#printing the subscriber information
print(jimTag[0].get_text()) 

#converting the value to integer for arithmetic calculation 
st =jimTag[0].get_text()
jim = st.split(" ")[0]
jimSub = int(jim.replace(",",""))
```

```
#printing the subscriber information
print(sbhTag[0].get_text())

#converting the value to integer for arithmetic calculation 
st =sbhTag[0].get_text()
sbh = st.split(" ")[0]
sbhSub= int(sbh.replace(",",""))
```

The two YouTube channels, "Jim Sterling" and "Super Bunnyhop" have 807552 and 418401 subscribers. 

Now that we have the subscriber count of each of the channels, we can now get the difference in subscriber number between the channels.

```
difSub = jimSub-sbhSub
print(difSub)
```

From the calculation, we get a difference of 389151 subscribers between "Jim Sterling" and "Super Bunnyhop". That is a significant gap in the difference between the channels compared to their respective subscription number. The above analysis in subscription difference states clearly that "Jim Sterling" is reaching more viewers regularly. As mentioned before, the channels cover a similar topic but their style of presenting videos are widely different. Also, the "Jim Sterling" is more consistent in producing videos every week compared to "Super Bunnyhop's" bi-weekly approach.

From the significant difference in subscriber number, it can be stated that the mentioned facts about presentation style and producing rate of the videos of the two channels have a great impact in terms of attracting subscribers.
